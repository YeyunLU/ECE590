```python
import numpy as np
def gaussian(m,n):
    mean=np.zeros(m)
    cov=np.identity(m)
    X = np.random.multivariate_normal(mean, cov, n)
    return X
```

```python
import matplotlib.pyplot as plt
import math
d = 40
samples =100
means = np.zeros(d)
stds = np.zeros(d)
r_hat = np.zeros(d)
for m in range(d):
    X = gaussian(m+1,samples)
    ridius = np.zeros(samples)
    for i in range(samples):
        ridius[i] = math.sqrt(np.sum(X[i,:]**2))
    means[m] = np.mean(ridius)
    stds[m] = np.std(ridius)
    r_hat[m] = math.sqrt(m+1)*stds[m]
m = np.linspace(1,d,d)
fig, ax = plt.subplots(1,2,figsize=(12, 5))
ax[0].plot(m,means)
ax[0].set_title('means')
ax[0].set_xlabel('dimension')
ax[1].plot(m,stds)
ax[1].set_title('standard deviation')
ax[1].set_xlabel('dimension')
plt.show()   
```
